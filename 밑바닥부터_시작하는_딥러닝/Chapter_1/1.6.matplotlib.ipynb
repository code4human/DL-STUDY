## 1.6.1. 단순한 그래프 그리기

## 1.6.2.pyplot 기능

cos 함수를 그려본다.

```python
#데이터 준비
x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

#그래프 그리기
plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle="--", label="cos") #cos 함수는 점선으로 그리기
plt.xlabel("x")   #x축 이름
plt.ylabel("y")   #y축 이름
plt.title('sin & cos')   #제목
plt.legend()
plt.show()
```
