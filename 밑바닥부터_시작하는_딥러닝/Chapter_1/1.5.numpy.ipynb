## 1.5.2. 넘파이 배열 생성하기
넘파이 배열을 만들 때 np.array() 메서드를 이용한다. <br>
np.array()는 파이썬의 리스트를 인수로 받아 넘파이 라이브러리가 제공하는 특수한 형태의 배열(numpy.ndarray)을 반환한다.

```python
x = np.array([1.0, 2.0, 3.0])
print(x)
print(type(x))
```

## 1.5.3.넘파이의 산술 연산
원소 수가 같다면 산술 연산은 각 원소에 대해 행해진다. <br>
원소 수가 다르면 오류가 발생한다.<br>
'원소별' = element-wise, '원소별 곱셈' = element-wise product <br><br>
넘파이 배열은 원소별 계산뿐 아니라 넘파이 배열과 수치 하나(스칼라값)의 조합으로 된 산술 연산도 수행한다.<br>
이 경우 스칼라값과의 계산이 넘파이 배열의 원소별로 한 번씩 수행된다. 이 기능을 <b>브로드캐스트</b>라고 한다.

```python
x = np.array([1.0, 2.0, 3.0])   
y = np.array([2.0, 4.0, 6.0])  
print(x+y)
print(x-y)
print(x*y)
print(x/y)
print(x/2.0)
```

## 1.5.4.넘파이의 N차원 배열
넘파이 배열(np.array)는 다차원 배열도 작성 가능하다.<br>
수학에서는 1차원 배열을 벡터(vector), 2차원 배열은 행렬(matrix)이라고 부른다.<br>
또 벡터와 행렬을 일반화한 것을 텐서(tensor)라고 한다.<br>
이 책에서는 2차원 배열을 '행렬', 3차원 배열을 '다차원 배열'이라고 명시한다.

```python
#2차원 배열(행렬)
A = np.array([ [1,2], [3,4] ])
print(A)
print(A.shape)  #행렬의 형상, 즉 차원의 크기(원소 수) (2X2)
print(A.dtype)  #행령에 담긴 원소의 자료형
```

형상이 같은 행렬끼리면 행렬의 산술 연산도 대응하는 원소별로 계산된다. (배열과 동일) <br>
행렬과 스칼라값의 산술 연산도 가능하다. 이때도 배열과 마찬가지로 브로드캐스트 기능이 작동한다.

```python
print(A)
print(A*10)
```

## 1.5.5.브로드캐스트

## 1.5.6.원소접근
원소의 인덱스는 0부터 시작한다.

```python
X = np.array([ [51,55],[14,19],[0,4] ])
print(X)
print(X[0])
print(X[0][1])
```

for문으로도 각 원소에 접근할 수 있다.

```python
for row in X:
    print(row)
```

인덱스를 배열로 지정해 한 번에 여러 원소에 접근할 수도 있다.

```python
X = X.flatten()   #X를 1차원 배열로 변환(평탄화)
print(X)
X[np.array([0,2,4])]   #인덱스가 0,2,4인 원소 얻기
```

이 기법을 응용하면 특정 조건을 만족하는 원소만 얻을 수 있다.

```python
#X에서 15이상인 값만 구하기
print(X > 15)    #X>15의 결과는 bool배열
print(X[X>15])   #배열 X에서 True에 해당하는 원소만 꺼내기
```

파이썬 같은 동적 언어(인터프리터 언어)는 C나 C++같은 정적 언어(컴파일 언어)보다 처리 속도가 늦다.<br>
실제로 무거운 작업을 할 때는 C/C++로 작성한 프로그램을 쓰는 편이 좋다.<br>
그래서 파이썬에서 빠른 성능이 요구될 경우 해당 부분을 C/C++로 구현하곤 한다.<br>
그때 파이썬은 C/C++로 쓰인 프로그램을 호출해주는, 이른바 '중개자'같은 역할을 한다.<br>
넘파이도 주된 처리는 C/C++로 구현돼있다. 그래서 성능을 해치지 않으면서 파이썬의 편리한 문법을 사용할 수 있다.
